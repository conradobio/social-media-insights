```python
# Initialize the ApifyClient with your API token
client = ApifyClient(apify_token)
```

```python
# https://apify.com/apify/instagram-scraper/api
# INPUTS - https://apify.com/apify/instagram-scraper/input-schema#directUrls
# Instagram Scraper allows you to scrape posts from a user's profile page, hashtag page, or place. 
# When a link to an Instagram post is provided, it can scrape Instagram comments.

run_input = {
    "directUrls": [
        "https://www.instagram.com/ic.campinas/"
        #"https://www.instagram.com/ic.saobernardo/"
    ],
    "proxy": {
        "useApifyProxy": True,
        "apifyProxyGroups": [
            "RESIDENTIAL"
        ]
    },
    "resultsLimit": 200,
    "resultsType": "posts", #details posts
    "searchLimit": 1,
    "searchType": "hashtag"
}
```

```python
# Run the actor and wait for it to finish
run = client.actor("apify/instagram-scraper").call(run_input=run_input)
```

```python
# Fetch and print actor results from the run's dataset (if there are any)
items = []
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    items.append(item)
```
